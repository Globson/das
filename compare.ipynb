# Exemplo

```python
df_example = pd.DataFrame(
    {
        "Idade": [21, 24, 25, 30, 34, 33, 38, 41, 39],
        "Altura (cm)": [
            160,
            154,
            158,
            170,
            169,
            176,
            183,
            190,
            180,
        ],
        "Peso (kg)": [
            50.55,
            60.6,
            48.8,
            76.8,
            54.7,
            67.9,
            79.0,
            80.6,
            83.1,
        ],
    }
)

df_example
```

```python
QIs = list(df_example.columns)
k_elbow = 0
for QI in QIs:
    k_elbow += findKElbow(df_example, QI)
k_elbow = round(k_elbow / len(QIs))
df_anon_separatiz = anonymizeSeparatiz(df_example, QIs, k_elbow, type="float")
df_anon_separatiz
```
